```python
import pandas as pd

drug_df = pd.read_csv('./dataset/drugs.csv')
drug_df
```

```python
a = drug_df.loc[:,'Na_to_K'].reset_index()
a
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
b = std.fit_transform(a[['Na_to_K']])
a['Na_to_K'] = b
a
```

```python
c_a = a['Na_to_K'] >= -1.96
c_b = a['Na_to_K'] <= 1.96

cond = c_a & c_b
drug_df = drug_df.iloc[a[cond].index].reset_index(drop=True)
drug_df
```

```python
drug = drug_df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

drug_encoder = LabelEncoder()
targets = drug_encoder.fit_transform(drug_df['Drug'].tolist())
drug['Drug'] = targets

sex_encoder = LabelEncoder()
targets = sex_encoder.fit_transform(drug_df['Sex'].tolist())
drug['Sex'] = targets

bp_encoder = LabelEncoder()
targets = bp_encoder.fit_transform(drug_df['BP'].tolist())
drug['BP'] = targets

ch_encoder = LabelEncoder()
targets = ch_encoder.fit_transform(drug_df['Cholesterol'].tolist())
drug['Cholesterol'] = targets
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd


features = drug.iloc[:,:-1]
targets = drug.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(features, targets)


```
