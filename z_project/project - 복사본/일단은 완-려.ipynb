```python
import pandas as pd


tag = pd.read_csv('./datasets/tbl_wishlist_tag.csv')
wish = pd.read_csv('./datasets/tbl_wishlist.csv')
member = pd.read_csv('./datasets/tbl_member.csv')
member_info = pd.read_csv('./datasets/tbl_member_favorite_category.csv')
```

```python
member_info

category_map = {
    1: '카테고리_취미',
    2: '카테고리_문화',
    3: '카테고리_액티비티',
    4: '카테고리_음식',
    5: '카테고리_여행',
    6: '카테고리_발전',
    7: '카테고리_동네모임',
    8: '카테고리_데이트',
    9: '카테고리_투자',
    10: '카테고리_언어',
    11: '카테고리_공부',
    12: '카테고리_축제',
    13: '카테고리_기타',
}
member_info['category_id'] = member_info['category_id'].map(category_map)
member_info
```

```python
columns = ['id', 'member_id', 'category_id', 'wishlist_content']

wish = wish[columns]
wish

columns = ['wishlist_id', 'tag_name',]

tag = tag[columns]
tag
```

```python
# tag.wishlist_id.duplicated().sum()

aa = member_info.groupby('member_id')['category_id'].apply(lambda x: ' '.join(x)).reset_index()
aa = pd.merge(wish, aa, left_on='member_id', right_on='member_id', how='left')
aa['category_id_y'] = aa['category_id_y'].fillna('')

# wish = merged_df
# wish
aa
```

```python
# tag.wishlist_id.duplicated().sum()

tag_concat = tag.groupby('wishlist_id')['tag_name'].apply(lambda x: ' '.join(x)).reset_index()
aa = pd.merge(aa, tag_concat, left_on='id', right_on='wishlist_id', how='left')
aa['tag_name'] = aa['tag_name'].fillna('')

wish = aa
wish
```

```python
columns = [ 'member_id', 'category_id_x', 'wishlist_content','category_id_y','tag_name']

wish = wish[columns]
wish['wishlist_content'] = wish['wishlist_content'] + ' ' +wish['tag_name'] + ' ' + wish['category_id_y']

columns = [ 'member_id', 'category_id_x', 'wishlist_content']

wish = wish[columns]
wish
```

```python
category_map = {
    1: '카테고리_취미',
    2: '카테고리_문화',
    3: '카테고리_액티비티',
    4: '카테고리_음식',
    5: '카테고리_여행',
    6: '카테고리_발전',
    7: '카테고리_동네모임',
    8: '카테고리_데이트',
    9: '카테고리_투자',
    10: '카테고리_언어',
    11: '카테고리_공부',
    12: '카테고리_축제',
    13: '카테고리_기타',
}
wish['category_id_x'] = wish['category_id_x'].map(category_map)
wish['wishlist_content'] = wish['wishlist_content'] + ' ' + wish['category_id_x']

columns = [ 'member_id', 'wishlist_content']

wish = wish[columns]
merged_df = wish
# # member_id를 기준으로 wishlist_content 합치기
merged_df = wish.groupby('member_id')['wishlist_content'].apply(lambda x: ' '.join(x)).reset_index()
merged_df
# # 각 wishlist_content를 단어로 분할하여 합치기
# merged_df['wishlist_content'] = merged_df['wishlist_content'].apply(lambda x: ' '.join(x.split()))
# print(merged_df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_v = CountVectorizer()
count_metrix = count_v.fit_transform(result_df)
# print(count_metrix)

c_s = cosine_similarity(count_metrix)
print(c_s)
print(c_s.shape)
```
