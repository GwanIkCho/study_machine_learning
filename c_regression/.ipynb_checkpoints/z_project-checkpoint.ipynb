```python
import pandas as pd

c_df = pd.read_csv('./datasets/card_clients.csv')
c_df
```

```python
c_df = c_df.drop(labels=['card_clients'], axis=1)
c_df = c_df.drop_duplicates()
c_df['LIMIT_BAL'] = c_df.pop('LIMIT_BAL')
```

```python
import pandas as pd



# 청구액과 관련된 새로운 특성 생성
c_df['Total_BILL_AMT'] = (c_df['BILL_AMT1'] * 1.3852) + \
                            (c_df['BILL_AMT2'] * (-0.0941)) + \
                            (c_df['BILL_AMT3'] * (-0.0991)) + \
                            (c_df['BILL_AMT4'] * 1.1864) + \
                            (c_df['BILL_AMT5'] * 1.3044) + \
                            (c_df['BILL_AMT6'] * 1.0157)

# 상환액과 관련된 새로운 특성 생성
c_df['Total_PAY_AMT'] = (c_df['PAY_AMT1'] * 1.4092) + \
                           (c_df['PAY_AMT2'] * 1.2547) + \
                           (c_df['PAY_AMT3'] * 1.2558) + \
                           (c_df['PAY_AMT4'] * 1.4396) + \
                           (c_df['PAY_AMT5'] * 1.7912) + \
                           (c_df['PAY_AMT6'] * 1.6771)

# PAY와 관련된 새로운 특성 생성
c_df['Total_PAY'] = (c_df['PAY_0'] * -9446.0367) + \
                       (c_df['PAY_2'] * -13140) + \
                       (c_df['PAY_3'] * -7145.1910) + \
                       (c_df['PAY_4'] * -4477.0357) + \
                       (c_df['PAY_5'] * -3042.3526) + \
                       (c_df['PAY_6'] * -10720)

# 결과 확인



c_df = c_df.drop(columns=['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'])
c_df = c_df.drop(columns=['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6'])
c_df = c_df.drop(columns=['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'])
c_df['LIMIT_BAL'] = c_df.pop('LIMIT_BAL')
```

```python
from sklearn.preprocessing import PowerTransformer

# box-cox: 양수만 가능
# yeo-johnson: 음수, 0 포함 가능, default

# ptf = PowerTransformer(method='box-cox', standardize=False)
ptf = PowerTransformer(standardize=False)
result = ptf.fit_transform(c_df[['LIMIT_BAL']])
c_df['LIMIT_BAL'] = result

result = ptf.fit_transform(c_df[['AGE']])
c_df['AGE'] = result

result = ptf.fit_transform(c_df[['Total_PAY_AMT']])
c_df['Total_PAY_AMT'] = result

result = ptf.fit_transform(c_df[['Total_PAY']])
c_df['Total_PAY'] = result

```

```python
# from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = c_df.iloc[:, :-1], c_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train.values, y_train.values)
prediction = l_r.predict(X_test.values)
get_evaluation(y_test.values, prediction)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(features):
    vif = pd.DataFrame()
    vif['vif_score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
    # print('MSE: {:.4f}, RMSE: {:.4f}, R2: {:.4f}'\
    #   .format(MSE, RMSE,  R2))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

features, targets = c_df.iloc[:, :-1], c_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train.values, y_train.values)
prediction = l_r.predict(X_test.values)
get_evaluation(y_test.values, prediction)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

features, targets = c_df.iloc[:, :-1], c_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

dt_r = DecisionTreeRegressor(random_state=124)
rf_r = RandomForestRegressor(random_state=124, n_estimators=1000)
gb_r = GradientBoostingRegressor(random_state=124)
xgb_r = XGBRegressor()
lgb_r = LGBMRegressor(n_estimators=100)

models = [dt_r, rf_r, gb_r, xgb_r, lgb_r]
for model in models:
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    print(model.__class__.__name__)
    get_evaluation(y_test, prediction)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
from sklearn.model_selection import KFold

features, targets = c_df.iloc[:, :-1], c_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

lgbm = LGBMRegressor()

lgbm_parameters = {
    'lgbm__n_estimators': [150,200,250,300],
    'lgbm__learning_rate': [0.3, 0.5,0.7, 1],
    'n_estimators': [10, 50, 100, 500, 1000]
}
kfold = KFold(n_splits=10, random_state=124, shuffle=True)
g_lgbm = GridSearchCV(lgbm, param_grid=lgbm_parameters,scoring='r2', cv=kfold, refit=True, return_train_score=True, n_jobs=-1)

g_lgbm.fit(X_train, y_train)
```

```python
lgbm_result_df = pd.DataFrame(g_lgbm.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(lgbm_result_df)
display(g_lgbm.best_estimator_)
```
