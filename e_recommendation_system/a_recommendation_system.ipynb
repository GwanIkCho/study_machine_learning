```python
import numpy as np

v1 = np.array([0,0,1])
v2 = np.array([0,1,1])
```

```python
from numpy import dot
from numpy.linalg import norm

def get_cosine_similarity(A,B):
    return dot(A,B) / (norm(A) * norm(B))
```

```python
get_cosine_similarity(v1,v2)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

v1 = np.array([[0,0,1]])
v2 = np.array([[0,1,1]])
cosine_similarity(v1,v2)
```

```python
import numpy as np
import pandas as pd

m_df = pd.read_csv('./datasets/movie_dataset.csv')
m_df
```

```python
m_df.info()
```

```python
columns = ['genres','keywords','overview','tagline','cast','director',]
pre_m_df = m_df[columns].fillna('')
pre_m_df.isna().sum()
```

```python
columns = ['genres','keywords','overview','tagline','cast','director',]
features = pre_m_df.loc[:,columns]

def concatenate(features):
    features.genres + ' ' +  features.keywords + ' ' +features.overview + ' ' +features.tagline +' ' + features.cast + ' ' +features.director

result_df = concatenate(features)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_v = CountVectorizer()
count_ma = count_v.fit_transform(result_df)
print(count_ma)


```
